```python
BENEFICIARY_COST = pd.read_csv("../data/2018 Beneficiary Cost file.csv")
PLAN_INFORMATION = pd.read_csv("../data/2018 Plan information.csv")
ZIP_STATE_COUNTY = pd.read_csv("../data/2018 Zip State County.csv")
NCPDP = pd.read_csv("../data/2018NCPDPFILE_CLEAN.csv")
GEO_LOC = pd.read_csv("../data/2018 Geographic locator file.csv")
FORMULARY_DATA = pd.read_csv("../data/FORMULARY_DATA.csv")
NDC_DATA = pd.read_csv("../data/NDC_DRUG_FILE_LIST_UTF8.csv")
```

```python
user_zip = input("Enter your zipcode: ").lstrip("0")
user_state = ZIP_STATE_COUNTY[ZIP_STATE_COUNTY['"ZIPCODE"']\
                    .isin([user_zip])]['"STATENAME"'].values[0].strip()
```

```python
user_pdp_region_code = GEO_LOC[GEO_LOC['STATENAME']\
                    .isin([user_state])]['PDP_REGION_CODE'].values[0]
user_pdp_region_code
```

```python
F_IDS_TO_RANK = PLAN_INFORMATION[PLAN_INFORMATION['PDP_REGION_CODE']\
                 .isin([str(user_pdp_region_code)]) &\
                PLAN_INFORMATION['CONTRACT_ID']\
                 .str.match(r'\bS\w+')]
F_IDS_TO_RANK
```

```python
num_meds = int(input("Enter how many drugs you wish to search: "))
user_generics = [(input("Enter a generic drug name: "),\
                  int(input("Enter a desired strength in mg: ")))\
                 for i in range(num_meds)]
```

```python
NDC_DATA['PRODUCTNDC_CLEAN'] = NDC_DATA['PRODUCTNDC'].apply(lambda s: s.replace("-",""))
user_ndc_output = {}
for drug,strength in user_generics:
    if strength == 0 or type(strength) != int:
        user_ndc_output[(drug,strength)] = list(NDC_DATA\
        [NDC_DATA['NONPROPRIETARYNAME (GENERIC)']\
         .str.contains(drug,case=False,regex=False).fillna(False)]\
        ['PRODUCTNDC_CLEAN'].values)
    else:
        user_ndc_output[(drug,strength)] = list(NDC_DATA\
        [NDC_DATA['NONPROPRIETARYNAME (GENERIC)']\
         .str.contains(drug,case=False,regex=False).fillna(False)&\
         NDC_DATA['ACTIVE_NUMERATOR_STRENGTH']\
         .isin([str(strength)]).fillna(0)]\
        ['PRODUCTNDC_CLEAN'].values)
user_ndc_output
```

```python
FORMULARY_DATA['NDC_CLEAN'] = FORMULARY_DATA['NDC']\
   .apply(lambda s: ("0"*(11-len(str(s)))+str(s))[1:-2] if len(str(s)) != 11 else s)
```

```python
# NDC_DATA[NDC_DATA['PRODUCTNDC_CLEAN'].isin(['00028147'])]
# FORMULARY_DATA[FORMULARY_DATA['NDC_CLEAN'].isin(["00028147"])]
```

```python
user_ndc_pa_tlv = {drug:[] for drug in user_generics}

for drug in user_generics:
    ndc = random.choice(user_ndc_output[drug])
    user_ndc_pa_tlv[drug]+=[
        ndc,
        FORMULARY_DATA[\
            FORMULARY_DATA['FORMULARY_ID']\
                   .isin(F_IDS_TO_RANK['FORMULARY_ID'].values) &\
            FORMULARY_DATA['NDC_CLEAN']\
                   .str.contains(ndc)]\
        [['PRIOR_AUTHORIZATION_YN','TIER_LEVEL_VALUE']]
    ]
user_ndc_pa_tlv
```
